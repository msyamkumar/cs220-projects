```python
def get_stat_total(pkmn):
    #Return sum of all stats(HP,Attack,Defense,Sp. Atk,Sp. Def, Speed) 
    sum = 0
    for stat in ['HP', 'Attack', 'Defense', 'Sp_Atk', 'Sp_Def', 'Speed']:
        func_name = "get_" + stat.lower()
        # func_name = "project.get_" + stat.lower()
        func = getattr(project, func_name)
        sum += func(pkmn)
    return sum  
```

```python
def simple_battle(pkmn1, pkmn2):
    if abs(get_stat_total(pkmn1) - get_stat_total(pkmn2)) < 20:
        winner = 'Draw'
    elif get_stat_total(pkmn1) > get_stat_total(pkmn2):
        winner = pkmn1
    else:
        winner = pkmn2
    return winner   
```

```python
def most_damage(attacker, defender):
    py_move = 10 * project.get_attack(attacker)/project.get_defense(defender)
    sp_move = 10 * project.get_sp_atk(attacker)/project.get_sp_def(defender)
    move = py_move if py_move >= sp_move else sp_move
    return move
#     if py_move >= sp_move:
#         return py_move
#     else:
#         return sp_move
```

```python
import math
def num_hits(attacker, defender):
    return math.ceil(project.get_hp(defender)/most_damage(attacker, defender))
```

```python
def battle(pkmn1, pkmn2):
    # Return the name of the pkmn that can take more hits from the other
    # pkmn. If both pkmn faint within the same number of moves, return the
    # string 'Draw'
    r2 = num_hits(pkmn1, pkmn2)
    r1 = num_hits(pkmn2, pkmn1)
    if r1 == r2:
        speed1 = project.get_speed(pkmn1)
        speed2 = project.get_speed(pkmn2)
        if speed1 == speed2:
            return 'Draw'
        elif speed1 > speed2:
            return pkmn1
        else:
            return pkmn2
            
    elif r1 > r2:
        return pkmn1
    else:
        return pkmn2
```
